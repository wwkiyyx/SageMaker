```python
region = image.crop((30,10,50,25))
region.show()
```

```python
region = region.transpose(PIL.Image.Transpose.ROTATE_90)
region.show()
```

```python
image.paste(region, (0,0,15,20))
image.show()
```

```python
black = PIL.Image.new("RGBA",(30,40))
black.paste(region, (0,0,15,20))
black.paste(region, (15,0,30,20))
black.paste(region, (0,20,15,40))
black.paste(region, (15,20,30,40))
display(black)
```

```python
r,g,b,a=image.split()
r.show()
g.show()
b.show()
a.show()
```

```python
rgb = PIL.Image.merge("RGB", (r, g, b))
display(rgb)
```

```python
rgb.save("/home/studio-lab-user/sagemaker-studiolab-notebooks/images/launcher_button.jpg","JPEG")
```

```python
rgb.resize((100,50)).show()
```

```python
display(rgb.rotate(45))
```

```python
rgb.transpose(PIL.Image.Transpose.FLIP_LEFT_RIGHT).show()
rgb.transpose(PIL.Image.Transpose.FLIP_TOP_BOTTOM).show()
rgb.transpose(PIL.Image.Transpose.ROTATE_90).show()
rgb.transpose(PIL.Image.Transpose.ROTATE_180).show()
rgb.transpose(PIL.Image.Transpose.ROTATE_270).show()
```

```python
import PIL.ImageFilter
rgb.filter(PIL.ImageFilter.DETAIL).show()
```

```python
import PIL.ImageEnhance
enh = PIL.ImageEnhance.Contrast(image)
enh.enhance(1.3).show("30% more contrast")
```

```python
from PIL import Image, ImageDraw, ImageFont
font = ImageFont.truetype("/home/studio-lab-user/.conda/envs/default/lib/python3.9/site-packages/werkzeug/debug/shared/ubuntu.ttf", 48)
#/home/studio-lab-user/.conda/envs/default/lib/python3.9/site-packages/cv2/qt/fonts/DejaVuSans.ttf
im = Image.new("RGB", (200, 200), "white")
d = ImageDraw.Draw(im)
d.line(((0, 100), (200, 100)), "gray")
d.line(((100, 0), (100, 200)), "gray")
d.text((100, 100), "Quick", fill="black", anchor="ms", font=font)
im.show()
```
